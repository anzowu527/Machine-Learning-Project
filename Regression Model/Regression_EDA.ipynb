## Row-wise Observations

### Found duplicated rows

```python
import pandas as pd

# Group by all columns
grouped = train_x1.groupby(train_x1.columns.tolist())

# Initialize a dictionary to store indices of each group
group_indices = {}

# Iterate through groups and store indices
group_number = 0
for _, group in grouped:
    if len(group) > 1:  # Only consider groups with more than one row
        group_indices[group_number] = group.index.tolist()
        group_number += 1

# Print or use the group indices
print(group_indices)

# Initialize a dictionary to store Egap values for each group
grouped_egap_values = {}

for group_num, indices in group_indices.items():
    # Extract corresponding Egap values for the group of indices
    grouped_egap_values[group_num] = train_y.loc[indices, 'Egap'].tolist()

# Print or use the grouped Egap values
print(grouped_egap_values)
```

### Average their Egap and remove duplicates

```python
# Initialize a dictionary to store the average Egap for each group
average_egap_per_group = {}

for group_num, egap_values in grouped_egap_values.items():
    # Calculate the average Egap for the group
    average_egap_per_group[group_num] = sum(egap_values) / len(egap_values) if egap_values else 0

# Print or use the average Egap values per group
print(average_egap_per_group)
```

```python
train_y1= train_y.copy()
for group_num, indices in group_indices.items():
    # Get the average Egap for this group
    avg_egap = average_egap_per_group[group_num]

    # Update the Egap values for the rows in this group
    train_y1.loc[indices, 'Egap'] = avg_egap
```

```python
# Initialize an empty list to store the indices of rows to drop
rows_to_drop = []

# Iterate through the group_indices
for indices in group_indices.values():
    if len(indices) > 1:
        # Add all indices except the first one to the rows_to_drop list
        rows_to_drop.extend(indices[1:])

# Drop these rows from train_x1 and train_y1
train_x2 = train_x1.drop(rows_to_drop)
train_y2 = train_y1.drop(rows_to_drop)
train_y2 = train_y2.drop(['Id'], axis=1)

train_x2 = train_x2.reset_index(drop=True)
train_y2 = train_y2.reset_index(drop=True) #droped duplicates
```

### Finding highly correlated rows

```python
import xgboost as xgb

dtrain = xgb.DMatrix(train_x2, label=train_y2)

# Define XGBoost parameters
params = {
    'objective': 'reg:squarederror',  # Use 'binary:logistic' for classification
    'max_depth': 8,
    'min_child_weight': 2.1,
    'eta': 0.3,
    'subsample': 0.95,
    'colsample_bytree': 0.65
}

# Train the model (adjust num_boost_round as needed)
bst = xgb.train(params, dtrain, num_boost_round=10)
importance = bst.get_score(importance_type='gain')
sorted_importance = sorted(importance.items(), key=lambda x: x[1], reverse=True)

feature_importances_df = pd.DataFrame(sorted_importance, columns=['Feature', 'Importance'])

drop = feature_importances_df[feature_importances_df['Importance']<4]['Feature'].tolist()
train_x3 = train_x2.drop(columns=drop)
train_x3 # dropped least important features
test_x2 = test_x1.drop(columns=drop)
test_x2
```

```python
Ttrain_x3 = train_x3.T
# Assuming the rest of your code is the same
correlation_matrix = Ttrain_x3.corr().abs()
high_corr = (correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
             .stack()
             .sort_values(ascending=False))
highly_correlated_pairs = high_corr[high_corr > 0.99997]
print(highly_correlated_pairs)


correlated_pairs_list = list(highly_correlated_pairs.index)
#print(correlated_pairs_list)

merged_groups = []

for pair in correlated_pairs_list:
    merged = False
    for group in merged_groups:
        # Check if the pair shares any element with the current group
        if set(pair).intersection(set(group)):
            # Merge the pair with this group
            group.update(pair)
            merged = True
            break
    if not merged:
        # If the pair wasn't merged with any existing group, add as a new group
        merged_groups.append(set(pair))

# Convert sets back to lists (optional)
merged_groups = [list(group) for group in merged_groups]

grouped_egap_values = {}

for group_num, indices in enumerate(merged_groups):
    # Extract corresponding Egap values for the group of indices
    grouped_egap_values[group_num] = train_y2.loc[indices, 'Egap'].tolist()

# Print or use the grouped Egap values
print(grouped_egap_values)
# Initialize a dictionary to store the average Egap for each group
average_egap_per_group = {}

for group_num, egap_values in grouped_egap_values.items():
    # Calculate the average Egap for the group
    average_egap_per_group[group_num] = sum(egap_values) / len(egap_values) if egap_values else 0

# Print or use the average Egap values per group
print(average_egap_per_group)
```

```python
train_y3 = train_y2.copy()

# Update the Egap values for each group
for group_num, indices in enumerate(merged_groups):
    avg_egap = average_egap_per_group[group_num]
    train_y3.loc[indices, 'Egap'] = avg_egap

# Initialize an empty list to store the indices of rows to drop
rows_to_drop = []

# Corrected loop to build rows_to_drop
for _, indices in enumerate(merged_groups):
    if len(indices) > 1:
        # Add all indices except the first one to the rows_to_drop list
        rows_to_drop.extend(indices[1:])

# Drop these rows from train_x3 and train_y3
train_x4 = train_x3.drop(rows_to_drop)
train_y4 = train_y3.drop(rows_to_drop)

# Reset the indices
train_x4.reset_index(drop=True, inplace=True)
train_y4.reset_index(drop=True, inplace=True)
```
